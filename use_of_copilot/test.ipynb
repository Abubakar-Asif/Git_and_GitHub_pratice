```python
# Create a line plot with dummy data using matplotlib
import matplotlib.pyplot as plt
import numpy as np
# Generate some dummy data
x = np.linspace(0, 10, 100)
y = np.sin(x)
# Create the plot
plt.plot(x, y)
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Dummy Line Plot')
# Show the plot
plt.show ()
```

```python
import seaborn as sns
import plotly.express as px

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Create a scatter plot using Plotly
fig = px.scatter(
    titanic,
    x='age',
    y='fare',
    color='sex',
    size='pclass',
    title='Titanic Age vs Fare Scatter Plot'
)

# Show the plot
fig.show()
```
